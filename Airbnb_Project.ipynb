# 2- Data Cleaning:
- Handle missing values appropriately (e.g., imputation, deletion, etc.).
- Check for and handle any duplicate entries.
- Convert categorical variables into the appropriate data type if necessary.


```python
# Drop Drop Unnecceray columns:
df.drop(['Listing ID','Name','Host ID','Host Name','Host Response Rate','Street',
         'Neighbourhood cleansed', 'State','Country','latitude','longitude','Bathrooms','Bedrooms',
         'Amenities','Minimum nights','Maximum nights','Availability 365','Calendar last scraped','Last Review Date','Host total listings count'], axis=1, inplace=True)
```

```python
# Change columns name to help us to use it in coding... Ex: 'Room Type' it doesn't help us in coding otherwise 'Room_Type':
df.rename(columns={'Host Is Superhost':'Host_Is_Superhost', 'Host total listings count':'Host_total_listings_count', 'Property type':'Property_Type',
          'Room type':'Room_Type', 'Number of reviews':'Number_of_reviews', 'Review Scores Rating':'Review_Scores_Rating','Review Scores Accuracy':'Review_Scores_Accuracy',
          'Review Scores Cleanliness':'Review_Scores_Cleanliness', 'Review Scores Checkin':'Review_Scores_Checkin', 'Review Scores Communication':'Review_Scores_Communication',
          'Review Scores Location':'Review_Scores_Location', 'Review Scores Value':'Review_Scores_Value','Reviews per month':'Reviews_per_month'},inplace=True)
```

```python
df.columns
```

```python
df
```

```python
# Check the missing values in each columns:
df.isnull().sum()
```

```python
# Dispaly the count of the missing values:
df.isnull().sum().sum()
```

```python
# We prefer to use mean to fill the missing values in 'Accommodates' column:
mean_value = df['Accommodates'].mean() 
df['Accommodates'].fillna(value=mean_value,inplace=True)
```

```python
# Check the missing values in each columns after filling the missing values by mean:
df.isnull().sum()
```

```python
df
```

As we see in the 'Accommodate' column, aftre using mean to fill the missing values it shown float type of values.

```python
df.Accommodates = df.Accommodates.round()
```

```python
df
```

After round the values to the nearest digits, the values still float, so, to solve this problem we can convert the type to integer.

```python
df['Accommodates'] = df['Accommodates'].astype(int)
```

```python
df
```

```python
df.duplicated()
```

```python
# Convert the 'Price' column to float data type to represent the price.
df['Price'] = df['Price'].astype(float)
```

```python
df
```

```python
df['Host_Is_Superhost'] = df['Host_Is_Superhost'].astype(int)
```

We convert categorical columns 'Host_Is_Superhost' that has values = 'False' to numerical columns = 0

```python
df.head()
```

```python
city_reviews = df.groupby(['City'])['Number_of_reviews'].max().reset_index()
city_reviews
```

```python
city = city_reviews['City']
reviews = city_reviews['Number_of_reviews']
fig = plt.figure(figsize = (10,5))

plt.bar(city, reviews, color='orange', width=0.5)
plt.xlabel('City')
plt.ylabel('Number_of_reviews')
plt.title('Number of Reviews by City')
plt.show()
```

```python
price_reviews = df.groupby(['Price'])['Number_of_reviews'].max().reset_index()
price_reviews
```

```python
price_list = price_reviews['Price']
reviews = price_reviews['Number_of_reviews']
fig = plt.figure(figsize=(10,5))

plt.scatter(price_list, reviews)
plt.xlabel('Price')
plt.ylabel('Number_of_reviews')
plt.title('The number of reviews based on the price')
plt.show()
```

From the above visualization we can say that the most number of people like to stay in less price and their reviews are higher in those areas.

**- Which hosts are the busiest and why?**

```python
busy_hosts = df.groupby(['Room_Type'])['Number_of_reviews'].max().reset_index()
busy_hosts
```

```python
room_type = busy_hosts['Room_Type']
reviews = busy_hosts['Number_of_reviews']
fig = plt.figure(figsize=(10,5))

plt.bar(room_type, reviews, width=0.3)
plt.xlabel('Room_Type')
plt.ylabel('Number_of_reviews')
plt.title('The most busiest room type in terms of reviews')
plt.show()
```

As we see in the graph, people prefer hotel room most other types of rooms.

```python
price_list = price_reviews['Price']
fig = plt.figure(figsize=(10,5))

plt.boxplot(price_list)
plt.xlabel('Price')
plt.title('Listing Price')
plt.show()
```

```python
def plot_hist(data1, data2):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,6))
    sns.histplot(data1, ax=ax1, kde=True)
    sns.histplot(data2, ax=ax2, kde=True);
```

```python
price_list = df["Review_Scores_Value"]
review_scores = df["Review_Scores_Cleanliness"]

print(f'Price: {price_list.min()} to {price_list.max()}')
print(f'Review Scores: {review_scores.min()} to {review_scores.max()}')

plot_hist(price_list, review_scores);
```

```python
from sklearn.preprocessing import MinMaxScaler

df[["Review_Scores_Value", "Review_Scores_Cleanliness"]] = MinMaxScaler().fit_transform(df[["Review_Scores_Value", "Review_Scores_Cleanliness"]])

plot_hist(df["Review_Scores_Cleanliness"], df["Review_Scores_Value"])
```
